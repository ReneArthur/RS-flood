```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Using the water height dataset 
df = pd.read_csv('rio_valores.csv').drop(columns=['Unnamed: 0'])
```

```python
# String Date to DateTime (Year format not relevant)
df['DIAS'] = pd.to_datetime(df["DIAS"], format=f"%d/%m/")

# Mean River Water Height
newDF = df.groupby(["DIAS"])['VALOR'].mean()
newDF
```

```python
# Plot
newDF.plot(kind='line', ylabel='Water Height (cm)', xlabel="Days (APR/MAY)")
```
